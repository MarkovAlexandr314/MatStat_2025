```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = np.linspace(-1, 3, 1000)

C1 = 0.767
C2 = 1.272
alpha = 0.05
n = 139
m = 1000
fig, ax = plt.subplots(1, 1)
a = np.linspace(0.1, 2, 100)

ax.plot(a, 1 - stats.f.cdf(C2, n - 1, m - 1, scale=a) +
         stats.f.cdf(C1, n - 1, m - 1, scale=a), color="green")

ax.set_ylabel("Мощность критерия")
ax.grid()
```
